# Fibonacci Sequence

```python
def fibonacci(n):
    res = [0, 1]
    for i in range(2, n + 1):
        res.append(res[i - 1] + res[i - 2])
    return res

n = int(input())
fibonacci(n)
```

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.tag import pos_tag
import matplotlib.pyplot as plt
```

```python
hotel_booking_tokens = []
def filter_tokens(tokens):
    relevant_pos_tags = ['NN', 'NNS', 'JJ', 'VB', 'VBP']
    filtered_tokens = []
    tagged_tokens = pos_tag(tokens)
    for token, tag in tagged_tokens:
        if tag in relevant_pos_tags:
            filtered_tokens.append(token.lower())
    return filtered_tokens

```

```python
for text in usertexts:
    tokens = word_tokenize(text)
    filtered_tokens = filter_tokens(tokens)
    hotel_booking_tokens.extend(filtered_tokens)



word_freq_map = {}


for word in hotel_booking_tokens:
    if word in word_freq_map:
        word_freq_map[word] += 1
    else:
        word_freq_map[word] = 1


print("Word Frequency Map:")
print(word_freq_map)
```

```python
sorted_word_freq_map = dict(sorted(word_freq_map.items(), key=lambda item: item[1], reverse=True))


tokens = list(sorted_word_freq_map.keys())
frequencies = list(sorted_word_freq_map.values())


plt.figure(figsize=(10, 6))
plt.plot(tokens, frequencies, marker='o')
plt.xlabel('Tokens')
plt.ylabel('Frequency')
plt.title('Token Frequency Map (User)')
plt.xticks(rotation=90)
plt.show()
```

```python
import pandas as pd
from nltk.tokenize import word_tokenize
from collections import Counter
from prefixspan import PrefixSpan
```

```python
# Load the conversation history into a DataFrame
conversation_history = ["User message 1", "Chatbot response 1", "User message 2", "Chatbot response 2", ...]
df = pd.DataFrame(conversation_history, columns=['text'])

# Preprocess the data
df['text'] = df['text'].str.lower().str.replace('[^\w\s]', '')

# Tokenize the sentences
df['tokens'] = df['text'].apply(word_tokenize)

# Analyze the frequency of words and phrases
word_freq = Counter()
for tokens in df['tokens']:
    word_freq.update(tokens)

# Print the most common words
print(word_freq.most_common(10))
```

```python
# Identify sequences of user messages and chatbot responses
# This will depend on how your data is structured. If each row alternates between user and chatbot, you could simply pair up every two rows.
conversation_history = ["User message 1", "Chatbot response 1", "User message 2", "Chatbot response 2", ...]
pairs = list(zip(conversation_history[::2], conversation_history[1::2]))



# Identify common conversation patterns or interaction sequences
flat_pairs = [str(pair) for pair in pairs]
pair_counts = Counter(flat_pairs)

# Print the most common pairs
print(pair_counts.most_common(10))
# This could involve more complex NLP techniques like sequence pattern mining or machine learning.

sequences = [[str(pair) for pair in conversation] for conversation in pairs]

# Initialize PrefixSpan with your sequences
ps = PrefixSpan(sequences)

# Find frequent patterns
frequent_patterns = ps.frequent(10)  # Find patterns that occur at least 10 times

# Print the frequent patterns
for pattern in frequent_patterns:
    print(f"Pattern: {pattern[1]}, Frequency: {pattern[0]}")
```
